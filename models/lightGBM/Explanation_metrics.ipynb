### 2. Classification Report

#### **各指標の説明**

##### **Precision (適合率)**
- **定義**: Positiveと予測されたサンプルのうち、どれだけが正しいかを示します。
- **クラスごとの計算**:
  - **クラス0 (非生存者)**:  
    - 計算式:  
      ```
      Precision = 96 / (96 + 9) = 0.91 (91%)
      ```
  - **クラス1 (生存者)**:  
    - 計算式:  
      ```
      Precision = 59 / (59 + 15) = 0.87 (87%)
      ```

##### **Recall (再現率)**
- **定義**: 実際にPositiveであるサンプルのうち、どれだけを正しく予測できたかを示します。
- **クラスごとの計算**:
  - **クラス0 (非生存者)**:  
    - 計算式:  
      ```
      Recall = 96 / (96 + 9) = 0.91 (91%)
      ```
  - **クラス1 (生存者)**:  
    - 計算式:  
      ```
      Recall = 59 / (59 + 15) = 0.80 (80%)
      ```

##### **F1-score**
- **定義**: PrecisionとRecallの調和平均であり、バランスの取れた指標。
- **クラスごとの計算**:
  - **クラス0 (非生存者)**:  
    - 計算式:  
      ```
      F1-score = 2 * 0.91 * 0.91 / (0.91 + 0.91) = 0.89 (89%)
      ```
  - **クラス1 (生存者)**:  
    - 計算式:  
      ```
      F1-score = 2 * 0.87 * 0.80 / (0.87 + 0.80) = 0.83 (83%)
      ```

##### **Macro Avg**
- **定義**: 各クラスの平均値。クラス間のバランスが取れているか確認するのに使用。
- **計算式**:  
      ```
      Macro Avg F1 = (0.89 + 0.83) / 2 = 0.86 (86%)
      ```

##### **Weighted Avg**
- **定義**: 各クラスのサンプル数に基づいて重み付けされた平均値。
- **目的**: 全体的なモデル性能を把握するための指標。
